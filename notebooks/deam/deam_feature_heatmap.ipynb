```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns   
```

```python
features_file = '../../data/combined_features_scores.csv'

# Read the averaged features
df = pd.read_csv(features_file)
```

```python
corr = df.corr(method="pearson")

# Focus only on correlations with valence_mean and arousal_mean
target_corr = corr[[" valence_mean", " arousal_mean"]].drop([" valence_mean", " arousal_mean"])
```

```python
plt.figure(figsize=(12, 50))
sns.heatmap(target_corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Feature Correlations with Valence & Arousal", fontsize=14)
plt.show()
```

```python
print("Top 10 features correlated with Valence:")
print(target_corr["valence_mean"].abs().sort_values(ascending=False).head(10))

print("\nTop 10 features correlated with Arousal:")
print(target_corr["arousal_mean"].abs().sort_values(ascending=False).head(10))
```
