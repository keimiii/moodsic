Calculate the correlation between each feature and the valence and arousal means to determine the more important features.

```python
import pandas as pd
```

```python
features_file = '../../data/combined_features_scores.csv'

# Read the averaged features
features_df = pd.read_csv(features_file)
```

```python
# Exclude target column from feature set
score_columns = ['song_id', ' valence_mean', ' valence_std', ' arousal_mean', ' arousal_std']
feature_cols = [col for col in features_df.columns if col not in score_columns]
```

### Valence
Start by computing the Pearson correlation scores for feature columns.

```python
# Compute correlations
correlations = {}
for col in feature_cols:
    correlations[col] = features_df[col].corr(features_df[" valence_mean"], method="pearson")

# Convert to DataFrame for readability
corr_df = pd.DataFrame.from_dict(correlations, orient="index", columns=["Pearson_corr"])
corr_df = corr_df.sort_values(by="Pearson_corr", ascending=False)

print(corr_df.head(10))
```

If we look at the combined_features_scores dataframe, we can see that most of the features are not normally distributed. So instead of just Pearson, we also check the Spearman correlation coefficients to perform feature analysis.

```python
# Compute correlations
correlations = {}
for col in feature_cols:
    correlations[col] = features_df[col].corr(features_df[" valence_mean"], method="spearman")

# Convert to DataFrame for readability
corr_df = pd.DataFrame.from_dict(correlations, orient="index", columns=["Spearman_corr"])
corr_df = corr_df.sort_values(by="Spearman_corr", ascending=False)

print(corr_df.head(15))

print(corr_df.tail(10))
```

```python
# Compute correlations
correlations = {}
for col in feature_cols:
    correlations[col] = features_df[col].corr(features_df[" valence_std"], method="spearman")

# Convert to DataFrame for readability
corr_df = pd.DataFrame.from_dict(correlations, orient="index", columns=["Spearman_corr"])
corr_df = corr_df.sort_values(by="Spearman_corr", ascending=False)

print(corr_df.head(15))

print(corr_df.tail(10))
```

## Arousal

# Feature Analysis with Random Forest + SHAP

```python
import pandas as pd
import shap
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
features_file = '../data/combined_features_scores.csv'

# Read the averaged features
df = pd.read_csv(features_file)

# Exclude target column from feature set
score_columns = ['song_id', ' valence_mean', ' valence_std', ' arousal_mean', ' arousal_std']
feature_cols = [col for col in df.columns if col not in score_columns]
```

```python
X = df.drop(columns=score_columns)
y_valence = df[" valence_mean"]
y_arousal = df[" arousal_mean"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_valence, test_size=0.2, random_state=42)
rf_valence = RandomForestRegressor(n_estimators=200, random_state=42)
rf_valence.fit(X_train, y_train)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_arousal, test_size=0.2, random_state=42)
rf_arousal = RandomForestRegressor(n_estimators=200, random_state=42)
rf_arousal.fit(X_train, y_train)
```

```python
explainer_valence = shap.TreeExplainer(rf_valence)
shap_values_valence = explainer_valence.shap_values(X)

explainer_arousal = shap.TreeExplainer(rf_arousal)
shap_values_arousal = explainer_arousal.shap_values(X)
```

```python
print("Top features for Valence:")
shap.summary_plot(shap_values_valence, X, plot_type="bar")

print("Top features for Arousal:")
shap.summary_plot(shap_values_arousal, X, plot_type="bar")
```

```python
valence_importances = pd.Series(rf_valence.feature_importances_, index=X.columns)
arousal_importances = pd.Series(rf_arousal.feature_importances_, index=X.columns)

combined = (valence_importances**2 + arousal_importances**2)**0.5
print("\nTop overlapping features for Valence & Arousal:")
print(combined.sort_values(ascending=False).head(10))
```

```python
# Convert the top 10 features to a list
top_10_features = combined.sort_values(ascending=False).head(10).index.tolist()
print("\nTop 10 features as a list:")
print(top_10_features)
```
